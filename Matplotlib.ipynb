```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = ["A","B","C","D","E"]
y = [4,7,7,9,5]
Z = [8,2,9,3,6]
```

```python
plt.xlabel("Name")
plt.ylabel("Number")
plt.title("Sample")
plt.bar(x,y, width = 0.5, color = "red")
plt.show()
```

```python
plt.xlabel("Name")
plt.ylabel("Number")
plt.title("Sample")
c = ["Red", "Yellow", "Blue", "Green", "Magenta"]
plt.bar(x,y, width = 0.5, color = c)
plt.show()
```

```python
plt.xlabel("Name")
plt.ylabel("Number")
plt.title("Sample")
c = ["Red", "Yellow", "Blue", "Green", "Magenta"]
plt.bar(x,y, width = 0.5, color = c, align = "edge", edgecolor = "Black", linewidth = 2, linestyle = ":", alpha = 0.3, label = "Names")
plt.legend()
plt.show()
```

```python
plt.xlabel("Name")
plt.ylabel("Number")
plt.title("Sample")

width = 0.2
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.bar(p,y, width, color = "Red", label = "Name1")
plt.bar(p1,Z, width, color = "Blue", label = "Name2")

plt.xticks(p+width/2, x)
plt.legend()
plt.show()
```

```python
plt.xlabel("Name")
plt.ylabel("Number")
plt.title("Sample")

width = 0.2
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.barh(p,y, width, color = "Red", label = "Name1")
plt.barh(p1,Z, width, color = "Blue", label = "Name2")

plt.xticks(p+width/2, x)
plt.legend()
plt.show()
```

```python
day = [1,2,3,4,5,6,7]
no =  [5,4,2,6,7,3,4]

plt.scatter(day,no, label = "Frequency")
plt.xlabel("Day")
plt.ylabel("No.")
plt.title("Weeks")
plt.legend()
plt.show()
```

```python
day = [1,2,3,4,5,6,7]
no =  [5,4,2,6,7,3,4]

plt.scatter(day,no, label = "Frequency", s = 37, marker = "*", color = "Red")
plt.xlabel("Day")
plt.ylabel("No.")
plt.title("Weeks")
plt.legend()
plt.show()
```

```python
x = np.random.randint(12,60,50)
print(x)
plt.hist(x, edgecolor = "Black")
plt.show()
```

```python
x = np.random.randint(12,60,50)
b = [10,20,30,40]
plt.hist(x,bins = b, edgecolor = "Black")
plt.show()
```

```python
x = np.random.randint(12,60,50)
print(x)
plt.hist(x, edgecolor = "Black", cumulative = -1)
plt.show()
```

```python
x = ["A","B","C","D","E"]
y = [5,7,7,9,5]
z = [5,2,9,3,6]
w = [5,6,7,8,9]
```

```python
width = 0.5

for i in range(len(y)):
    p[i] = y[i]+z[i]
    
plt.bar(x,y, width)
plt.bar(x,z, width, bottom = y)
plt.bar(x,w, width, bottom = p)
plt.show()
```

```python
x = ["A","B","C","D","E"]
y = [5,7,7,9,5]
```

```python
plt.pie(y, labels = x)
plt.legend()
plt.show()
```

```python
ex = [0.4,0.0,0.0,0.0,0.0]

plt.pie(y, labels = x, explode = ex, autopct = "%0.1f%%", radius = 1.5)
plt.show()
```
